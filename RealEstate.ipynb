```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df1 = pd.read_csv('house_prices.csv')
print("Shape: " , df1.shape)
df1.head()
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
df2 = df1.drop(['area_type','society','balcony','availability'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
# To handle the NA values in dataframe
df3 = df2.dropna()
df3.isnull().sum()
```

```python
# To handle BHK/Bedroom in size column
df3['bhk'] = df3['size'].apply(lambda x: int(x.split(' ')[0]))
df3.head()
```

```python
df3.total_sqft.unique()
```

```python
# To handle ranges in total_sqft column
def is_float(x): # Function to check for float value (range)
    try:
        float(x)
    except:
        return False
    return True

print("df3: (total_sqft column is unchanged) \n" , df3[~df3['total_sqft'].apply(is_float)].head() , "\n")

def convert(x): # Function to take average if range of two numbers is given
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None

df4 = df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert)

print("df4: (total_sqft column is changed)" , df4.head())
```

```python
# Adding a column price_per_sqft
df5 = df4.copy()
df5['price_per_sqft'] = (df5['price']*100000)/(df5['total_sqft'])
df5.head()
```

```python
print(df5['location'].unique())
print(df5['location'].unique().size)
```

```python
df5.location = df5.location.apply(lambda x: x.strip()) # to remove any space before or after location
location_stats = df5['location'].value_counts() # to count the number with same location
print(location_stats)
```

```python
print("No. of Locations :" , len(location_stats))
print("No of Locations less than 10 :" , len(location_stats[location_stats<=10]))
print("No of locations greater than 10 :" , len(location_stats[location_stats>10]))
```

```python
# To remove the locations which are less than 10
less = location_stats[location_stats<=10]
print(less ,"\n")
df5.location = df5.location.apply(lambda x: 'other' if x in less else x)
print("No of location >10 :" , len(df5.location.unique()))
```

```python
df5.head(10)
```

```python
# To remove outliers where no of bedrooms are more when sqft is less
df5[(df5.total_sqft/df5.bhk)<300].head()
df6 = df5[~((df5.total_sqft/df5.bhk)<300)]
df6.head()
```

```python
# To remove outliers where price per sqft is different from most
def remove(df):
    new_df = pd.DataFrame()
    for key, col in df.groupby('location'):
        mean = np.mean(col.price_per_sqft)
        std = np.std(col.price_per_sqft)
        reduced_df = col[(col.price_per_sqft>(mean - std)) & (col.price_per_sqft <= (mean + std))]
        new_df = pd.concat([new_df, reduced_df], ignore_index=True)
    return new_df
df7 = remove(df6)
df7.shape
```

```python
# To remove outlier where prices are different for same number of bedrooms
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices,axis='index')
remove_bhk_outliers(df7)
# df8 = df7.copy()
df7.shape
```

```python
# Removing Outliers w.r.t bathrooms
print("Values : " , df7.bath.unique())
print("\nMore than 10 bathrooms :\n" , df7[df7.bath>10])
print("\nNo. of Bathrooms > No. of Bedrooms+2\n" , df7[(df7.bath)>(df7.bhk+2)]) # Outliers where no. of bathrooms is greater than no. of bedrooms+2
df8 = df7[(df7.bath)<(df7.bhk+2)] # Removed the outliers
df8.shape
```

```python
# Removing unnecessary features
df9 = df8.drop(['size','price_per_sqft'],axis='columns')
df9.head(5)
```

```python
# Encoding Location column using pandas dummies for ML Model
dummies = pd.get_dummies(df9.location)
dummies = dummies.drop('other',axis='columns') # Removing the other column
dummies.head(3)
```

```python
# removing location column
df10 = pd.concat([df9.drop('location',axis='columns'),dummies],axis='columns')
df10.head()
```

```python
print("Final Data after pre-processing:")
df = df10.copy()
df.head()
```

```python
x = df.drop(['price'],axis='columns')
y = df.price
```

```python
# Input Column
x.head(3)
```

```python
# Output Columns
y.head(3)
```

```python
# ML Model -> LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)

from sklearn.linear_model import LinearRegression 
lr_model = LinearRegression()
lr_model.fit(x_train,y_train)
lr_model.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(LinearRegression(), x, y, cv=cv)
```

```python
import pickle 
with open('RealEstate.pickle','wb') as data:
    pickle.dump(lr_model,data)
```
